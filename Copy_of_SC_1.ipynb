```python
def union(a,b):
  #setUnion= a.union(b)
  setUnion = set()
  for x in a:
    setUnion.add(x)
  for y in b:
    setUnion.add(y)
  print('Set Union: ',setUnion)
```

```python
def intersection(a,b):
  #setIntersection= a.intersection(b)
  setIntersection = set()
  for x in a:
    if x not in b:
      setIntersection.add(x)
  print('Set Intersection: ',setIntersection)
```

```python
def complement(a,b):
  #print('Set A complement: ', b.difference(a))
  #print('Set B complement: ', a.difference(b))
  setUnion= a.union(b)
  compA = set()
  compB = set()
  for x in setUnion:
    if x not in a:
      compA.add(x)
  for y in setUnion:
    if y not in b:
      compB.add(y)
  print('Set A complement: ', compA)
  print('Set B complement: ', compB)
    
```

```python
def univ(a,b):
  uv = set()
  for x in a:
    uv.add(x)
  for y in b:
    uv.add(y)
  return uv
```

```python
def cont(a,b):
  for x in a:
    if not b.__contains__(x):
      return False
    elif a[x]<b[x]:
      return False
  return True
```

```python
def union1(a,b):
  union={}
  uv = univ(a,b)
  for x in uv:
    try:
      a[x]
      max = a[x]
    except:
      max = 0
    try:
      b[x]
      max = b[x] if b[x] > max else max
    except:
      max = 0
    union[x] = max
  return union
```

```python
def complement(a):
  ca = {}
  uv = univ(a,b)
  for x in uv:
    try:
      ca[x] = 1 - a[x]
    except:
      ca[x] = 1
  return ca
  
```

```python
def intersection(a,b):
  intersect={}
  for x in a:
    if b.__contains__(x):
      intersect[x] = a[x] if a[x] < b[x] else b[x]
  return intersect
```

```python
def demorgan1(a,b):
  RHS = union1(complement(a),complement(b))
  LHS = complement(intersection(a,b))
  return LHS == RHS
```

```python
def demorgan2(a,b):
  RHS = intersection(complement(a),complement(b))
  LHS = complement(union1(a,b))
  return LHS == RHS
```

```python
# Test Case 1
a = {1:0.2,2:0.3,3:0.8,4:1}
b = {1:0.3,2:0.2,3:0.5,4:0.8}
print("A containment B: "+str(cont(a,b)))
print("B containment A: "+str(cont(b,a)))
print("Universal: ",univ(a,b))
print("Union: ",union1(a,b))
print("Intersection: ",intersection(a,b))
print("A complement: ",complement(a))
print("B complement: ",complement(b))
print("Demorgan 1: ",demorgan1(a,b))
print("Demorgan 2: ",demorgan2(a,b))
```
